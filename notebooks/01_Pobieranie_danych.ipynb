Pobieram dane z endpoint'u `airport`

```python
airport_data = {}
failed_airport_requests = {}
auth_header = {'authorization': 'WpzDMZeeCq6tbPdsTHUX8W9mecuUVwXAnmcorefr'}
count_since_last_sleep = 0
time_since_last_sleep = 0
for airport_id in airport_ids:
    start = time()
    URI = f'https://api-datalab.coderslab.com/api/airport/{airport_id}'
    response = requests.get(URI, headers=auth_header)
    end = time()
    time_since_last_sleep += (end - start)
    if response.status_code == 200:
        airport_data[airport_id] = response.json()
        print(f'Fetching data for {airport_id} took {end - start}s.')
    else:
        failed_airport_requests[airport_id] = response
        print(f'Fetching data for airport {airport_id} failed with code: {response.status_code} and error {response.json()}.')
    count_since_last_sleep += 1
    if (count_since_last_sleep == 17) and (time_since_last_sleep > 1):
        print(f'Exceeded 16 requests in 1s, sleeping 1s.')
        sleep(1)
        time_since_last_sleep = 0
        count_since_last_sleep = 0
```

```python
airport_df = pd.DataFrame.from_dict(airport_data, orient='index')
airport_df
```

Pobieram dane dotyczące zarejestrowanej pogody na poszczególnych lotniskach - z endpointu `weather`

```python
import time
import requests
from datetime import date, timedelta, datetime
from dateutil.relativedelta import relativedelta

weather_data_full = []
failed_w_requests = {}

auth_header = {'authorization': 'WpzDMZeeCq6tbPdsTHUX8W9mecuUVwXAnmcorefr'}
count_since_last_sleep = 0
time_since_last_sleep = 0

start_date = date(2019, 1, 1)
end_date = date(2020, 3, 31)
current_date = start_date

while current_date <= end_date:
    formatted_month = current_date.strftime('%Y-%m')
    url = f'https://api-datalab.coderslab.com/api/airportWeather?date={formatted_month}'
    print(f"Pobieranie danych dla daty: {formatted_month}")
    response = requests.get(url, headers=auth_header)
    
    if response.status_code == 200:
        try:
            month_data = response.json()
            if month_data:
                weather_data_full.extend(month_data)
            else:
                print(f"Brak danych dla daty: {formatted_date}")
        except KeyError:
            print(f"Brak danych dla miesiąca: {formatted_month}")
    else:
        failed_w_requests[formatted_date] = response.status_code
    
    current_date += relativedelta(months=1)

    count_since_last_sleep += 1
    time_since_last_sleep += response.elapsed.total_seconds()
    if count_since_last_sleep == 17 and time_since_last_sleep > 1:
        print(f'Przekroczono 16 żądań w ciągu 1 sekundy, oczekiwanie 1 sekundy.')
        time.sleep(1)
        time_since_last_sleep = 0
        count_since_last_sleep = 0
```

```python
# zbieram wszystkie nazwy kolumn, by zbudować tabelę - niestety przy niektórych lotniskach brakuje niektórych kolumn

weather_data_cols = set()
for entry in weather_data_full:
    weather_data_cols.update(entry.keys())
```

```python
# buduję słownik i uzupełniam brakujace kolumny NaN przy lotniskach
weather_data_dict = {}
for record in weather_data_full:
    for col in weather_data_cols:
        if col not in weather_data_dict:
                weather_data_dict[col] = []
        if col in record:
            weather_data_dict[col].append(record[col])
        else:
            weather_data_dict[col].append(None)
```

```python
weather_data_df = pd.DataFrame(weather_data_dict)
weather_data_df
```

Zapisuję dataframe do pliku `airport_weather.csv`

```python
abs_file_path = f'{PARENT_DIR}\\data\\raw\\airport_weather.csv'

weather_data_df.to_csv(abs_file_path, sep=';', encoding='utf-8', index=False)
```

Pobieram dane dotyczące ruchu lotniczego - endpoint `flights`

```python
import time
import requests
from datetime import date, timedelta, datetime
from dateutil.relativedelta import relativedelta

flight_data = {}
failed_flight_requests = {}
auth_header = {'authorization': 'WpzDMZeeCq6tbPdsTHUX8W9mecuUVwXAnmcorefr'}
count_since_last_sleep = 0
time_since_last_sleep = 0

start_date = date(2019, 1, 1)
end_date = date(2020, 3, 31)
current_date = start_date

for airport_id in airport_ids:
    airport_data_dict = {}
    current_date = start_date
    while current_date <= end_date:
        formatted_date = current_date.strftime('%Y-%m')
        url = f'https://api-datalab.coderslab.com/api/flight?airportId={airport_id}&date={formatted_date}'
        print(f"Pobieranie danych dla lotniska {airport_id} i daty {formatted_date}")
        response = requests.get(url, headers=auth_header)
        
        if response.status_code == 200:
            data = response.json()
            airport_data_dict[formatted_date] = data
        else:
            failed_flight_requests[formatted_date] = response.status_code
        
        current_date += relativedelta(months=1)

        count_since_last_sleep += 1
        time_since_last_sleep += response.elapsed.total_seconds()
        if count_since_last_sleep == 17 and time_since_last_sleep > 1:
            print(f'Przekroczono 16 żądań w ciągu 1 sekundy, oczekiwanie 1 sekundy.')
            time.sleep(1)
            time_since_last_sleep = 0
            count_since_last_sleep = 0
    flight_data[airport_id] = airport_data_dict
```

```python
#niepotrzebnie dałam to od razu do słownika, ale nie chciało mi się przerabiać. Więc teraz przerabiam to na listę
flight_data_dict = flight_data
flight_data_list = []
for airport_id, airport_data_dict in flight_data_dict.items():
    for monthyear, month_data_list in airport_data_dict.items():
        for data in month_data_list:
            flight_data_list.append(data)
```

```python
flight_df = pd.DataFrame(flight_data_list)
```

Zapisuję dataframe do pliku `flight.csv`

```python
abs_file_path = f'{PARENT_DIR}\\data\\raw\\flight.csv'
flight_df.to_csv(abs_file_path, sep=';', encoding='utf-8', index=False)
```
